```python
bad_90_q["date of disbursement"].dt.to_period("Q") #CONVERTING THE DATE OF DISBURSEMENT TO QUARTERS
```

```python
bad_90_q['Quarters'] = bad_90_q["date of disbursement"].dt.to_period("Q") #ADDING THE NEW CREATED SERIES TO THE BAD_90_Q SUB DATA FRAME
```

```python
bad_90_q = bad_90_q.replace(np.nan, 0) #REPLACING THE MISSING VALUES WITH ZEROS
```

```python
raw_bad_groups = bad_90_q.groupby('Quarters') #GROUPING THE DATA ON THE QUARTERS COLUMN
# raw_bad_groups.groups
```

```python
quaters = {}
for key, group_dataframe in raw_bad_groups:  #LOOPING THROUGH THE KEYS AND VALUES IN THE QUARTERS
    quaters[str(key)] = {}
    quaters[str(key)]["data"] = group_dataframe #APPENDING THE VALUES FROM THE LOOP
    
    bd_group = group_dataframe[['bad90_m4', 'bad90_m5',
       'bad90_m6', 'bad90_m7', 'bad90_m8', 'bad90_m9', 'bad90_m10',
       'bad90_m11', 'bad90_m12', 'bad90_m13', 'bad90_m14', 'bad90_m15',
       'bad90_m16', 'bad90_m17', 'bad90_m18', 'bad90_m19', 'bad90_m20',
       'bad90_m21', 'bad90_m22', 'bad90_m23', 'bad90_m24', 'bad90_m25',
       'bad90_m26', 'bad90_m27', 'bad90_m28']].sum()
    
    ds_group = bad_90_q['disbursed'].sum()
    bad_rate2 = (bd_group/ds_group) #CALCULATION FOR BAD RATE
    
    quaters[str(key)]["bad_90_rate"] = bad_rate2 #ASSIGNING THE BAD RATE TO A NEW VARIABLE
    
quaters[str(key)]["data"]
```

```python
plt.figure(num=None, figsize=(20, 10), dpi=80, facecolor='w', edgecolor='k') 

for key, value in quaters.items(): #LOOPING THROUGH THE ITEMS IN QUARTERS
    
    plt.plot(quaters[key]["data"].columns[3:-1], quaters[key]["bad_90_rate"], label = key) #PLOTTING THE COLUMNS AGAISNT THE BADRATE
    plt.xticks(rotation = 30)

plt.legend() #DISPLAY THE KEYS
plt.xlabel('Bad definition')
plt.ylabel('Bad rate')
plt.show()
```
